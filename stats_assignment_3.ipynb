#### Problem Statement 1: Blood glucose levels for obese patients have a mean of 100 with a standard deviation of 15. A researcher thinks that a diet high in raw cornstarch will have a positive effect on blood glucose levels. A sample of 36 patients who have tried the raw cornstarch diet have a mean glucose level of 108. Test the hypothesis that the raw cornstarch had an effect or not.
we have ; <br>
$\mu $ = 100   <br>

$\sigma $ = 15 <br>

n=36  <br>

$bar X $ = 108 <br>

$H_{0} $ = Raw cornstrach has effect on obesity ($\mu $ =100 )   <br>

$H_{1} $ = Raw cornstrach has no any  effect on obesity ( $\mu $ !=100)   <br>

***************Will follow two tail test********************************* <br> 

$ Z_{test}  = \frac {(\bar X - \mu)} {\frac {\sigma} {\sqrt n}} $  <br>

$ Z_{test}  = \frac {(108 - 100)} {\frac {15} {\sqrt 36}} $ = 3.2 <br>

Now we will take significance level asd 95% hence ; <br>

for two tail test $\frac {\alpha} {2} $ = 0.025

$Z_{0.025} $ = -1.96 <br>

3.2 >  -1.96  <br>

$ Z_{test} $ does not lies int the given region hence rejects null hypothesis




#### Problem Statement 2: In one state, 52% of the voters are Republicans, and 48% are Democrats. In a second state, 47% of the voters are Republicans, and 53% are Democrats. Suppose a simple random sample of 100 voters are surveyed from each state. What is the probability that the survey will show a greater percentage of Republican voters in the second state than in the first state?
Note:- Whenever we will have percentage population value we will try tapply population proportion concept;   <br>
now we have; <br>
$n_{1}$ = 100   <br>
$n_{2}$ = 100   <br>
$\hat p_{1}$ = 0.52  <br>
$(1-\hat p_{1})$= 0.48   <br>
$\hat p_{2}$ 0.47    <br>
$(1-\hat p_{2})$ = 0.53   <br>


$H_{0} $ = value >0.52 (Greater percentage of republican voters in second states than first state) <br>

$H_{1} $ = value <=0.52 (Lesser percentage of republican voters in second states than first state) <br>


Applying Z-test for population propotion formula  ; <br>

$ Z_{test} $ = $\frac {(\mu_{1} - \mu_{2})- (\hat p_{1} -\hat p_{2})} {\frac {(\hat p_{1})(1-\hat p_{1})}{n_{1}} + \frac {(\hat p_{2})(1-\hat p_{2})}{n_{2}} }$   <br>

$ Z_{test} $ = $\frac {(0)- (0.52 -0.47)} {\frac {(0.52)(0.48)}{100} + \frac {(0.47)(0.53)}{100} }$  = -0.7082   <br>

$Z_{0.05} $ = -1.65   <br>


$ Z_{test} $   > $Z_{0.05} $    <br>

Accept null hypothesis  <br>

Result :- Greater percentage of republican voters in second states than first state
